```python
import rungekutta4 as rk4
import numpy as np
import matplotlib.pyplot as plt
```

```python

def a(x,r):
    return [r[2],r[3],-r[0]-(2*r[0]*r[1]),-r[1]-r[0]**2+r[1]**2]

V=rk4.Yx(a,2,[1,-0.49,0,0.1],rk4.K1,rk4.K2,rk4.K3,rk4.K4,0.005)

X=V[:,[0]]
Y=V[:,[1]]

plt.plot(X,Y)

plt.show()
```

```python
def HenonHeilles(x,y):
    return (1/2)*(x**2+y**2)+(x**2)*y-(y**3)/3
def paraboloide(x,y):
    return x**2+y**2

def graficador(xo,x1,f,n):
    x=np.linspace(xo,x1,n)
    xy=np.meshgrid(x,x)
    Z=f(xy[0],xy[1])
    plt.contour(xy[0],xy[1],Z,[0.05,0.16666,0.5],colors='blue')
    V=rk4.Yx(a,5,[0.15,-.2,0,-1],rk4.K1,rk4.K2,rk4.K3,rk4.K4,0.05)
    X=V[:,[0]]
    Y=V[:,[1]]
    print(X,Y)
    plt.plot(X,Y, color='red')
    plt.show()

graficador(-2.5,2.5,HenonHeilles,1000)
```

```python
def separatrices(y,V):
    return (y,(-((y**2/2-y**3/3)-V)/(1/2+y))**(1/2),-(-((y**2/2-y**3/3)-V)/(1/2+y))**(1/2))
y=np.linspace(-2,2,1000)
print(separatrices(-1/2-.1,1/6)[1])
plt.plot(separatrices(y,1/6-0.1)[1],y, color='red')
plt.plot(separatrices(y,1/6-0.1)[2],y, color='red')
plt.show()
```
